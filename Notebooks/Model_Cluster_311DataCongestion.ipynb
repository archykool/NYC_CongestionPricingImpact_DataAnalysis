```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
from shapely.wkt import loads
from sklearn.cluster import KMeans
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Step 1: Load your NYC ZIP code boundaries
geo = pd.read_csv('Modified_Zip_Code_Tabulation_Areas__MODZCTA__20250427.csv')
geo['geometry'] = geo['the_geom'].apply(loads)
geo_gdf = gpd.GeoDataFrame(geo, geometry='geometry', crs="EPSG:4326")
geo_gdf['MODZCTA'] = pd.to_numeric(geo_gdf['MODZCTA'], errors='coerce')

# Step 2: Prepare incident data
df_con['Incident Zip'] = pd.to_numeric(df_con['Incident Zip'], errors='coerce')
df_con['Created Date'] = pd.to_datetime(df_con['Created Date'], errors='coerce')
df_con['Year'] = df_con['Created Date'].dt.year

# Step 3: Function to cluster per year
def cluster_zip_by_year(year, df_con, geo_gdf):
    df_year = df_con[df_con['Year'] == year].dropna(subset=['Incident Zip']).copy()
    zip_counts = df_year.groupby('Incident Zip').size().reset_index(name='incident_count')
    if len(zip_counts) < 3:
        return None  # not enough data for 3 clusters

    kmeans = KMeans(n_clusters=3, random_state=0)
    zip_counts['cluster'] = kmeans.fit_predict(zip_counts[['incident_count']])

    cluster_centers = kmeans.cluster_centers_.flatten()
    sorted_clusters = sorted(range(3), key=lambda x: cluster_centers[x])
    mapping = {old: new for new, old in enumerate(sorted_clusters)}
    zip_counts['congestion_level'] = zip_counts['cluster'].map(mapping)

    geo_merged = geo_gdf.merge(zip_counts, left_on='MODZCTA', right_on='Incident Zip', how='left')
    return geo_merged

# Step 4: Cluster each year separately
geo_2023 = cluster_zip_by_year(2023, df_con, geo_gdf)
geo_2024 = cluster_zip_by_year(2024, df_con, geo_gdf)
geo_2025 = cluster_zip_by_year(2025, df_con, geo_gdf)

# Step 5: Set up side-by-side subplots
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=("2023", "2024", "2025"),
    specs=[[{'type': 'mapbox'}, {'type': 'mapbox'}, {'type': 'mapbox'}]]
)

# Common settings
mapbox_style = "carto-positron"
center_nyc = dict(lat=40.7128, lon=-74.0060)
zoom_nyc = 9

# Step 6: Add maps
for i, geo_year in enumerate([geo_2023, geo_2024, geo_2025], start=1):
    if geo_year is not None:
        fig.add_trace(
            go.Choroplethmapbox(
                geojson=geo_year.__geo_interface__,
                locations=geo_year.index,
                z=geo_year['congestion_level'],
                colorscale='viridis',
                showscale=False,
                marker_opacity=0.7,
                marker_line_width=0
            ),
            row=1, col=i
        )

# Step 7: Update layout
fig.update_layout(
    height=700,
    title_text="NYC Zip Code Congestion Clustering (2023-2025, Viridis Color)",
    mapbox_style=mapbox_style,
    mapbox_center=center_nyc,
    mapbox_zoom=zoom_nyc
)

# Update each mapbox individually
fig.update_layout(
    mapbox1=dict(style=mapbox_style, center=center_nyc, zoom=zoom_nyc),
    mapbox2=dict(style=mapbox_style, center=center_nyc, zoom=zoom_nyc),
    mapbox3=dict(style=mapbox_style, center=center_nyc, zoom=zoom_nyc)
)

fig.show()
```

```python
"""
NYC ZIP‑code congestion clustering
Global normalisation + single shared KMeans
Diagnostic print‑outs + fixed y‑axis + discrete colours
------------------------------------------------------------------
Replace the two file paths below (GEO_PATH, CON_PATH) if needed,
or comment out the CSV load block if `df_con` is already defined.
"""

# -----------------------------------------------------------------
# 0  Imports
# -----------------------------------------------------------------
import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.wkt import loads

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

import plotly.graph_objects as go
from plotly.subplots import make_subplots

# -----------------------------------------------------------------
# 1  Geometry (MODZCTA polygons)
# -----------------------------------------------------------------
GEO_PATH = "Modified_Zip_Code_Tabulation_Areas__MODZCTA__20250427.csv"
geo = pd.read_csv(GEO_PATH)
geo["geometry"] = geo["the_geom"].apply(loads)
geo_gdf = gpd.GeoDataFrame(geo, geometry="geometry", crs="EPSG:4326")
geo_gdf["MODZCTA"] = pd.to_numeric(geo_gdf["MODZCTA"], errors="coerce")

# -----------------------------------------------------------------
# 2  311‑style incident data
# -----------------------------------------------------------------


# -----------------------------------------------------------------
# 3  Clean & enrich
# -----------------------------------------------------------------
df_con["Created Date"] = pd.to_datetime(df_con["Created Date"], errors="coerce")
df_con["Incident Zip"] = pd.to_numeric(df_con["Incident Zip"], errors="coerce")
df_con = df_con.dropna(subset=["Created Date", "Incident Zip"])
df_con["Year"] = df_con["Created Date"].dt.year
df_con["Month"] = df_con["Created Date"].dt.month

# -----------------------------------------------------------------
# 4  Jan–Apr incident rate per (Year, ZIP)
# -----------------------------------------------------------------
JAN_APR = [1, 2, 3, 4]
jan_apr = df_con.query("Month in @JAN_APR")

totals = (
    jan_apr.groupby(["Year", "Incident Zip"])
    .size()
    .reset_index(name="total_incident_count")
)
months_seen = (
    jan_apr.groupby(["Year", "Incident Zip"])["Month"]
    .nunique()
    .reset_index(name="n_months")
)

zip_year_rates = totals.merge(months_seen)
zip_year_rates = zip_year_rates[zip_year_rates["n_months"] > 0]
zip_year_rates["rate"] = (
    zip_year_rates["total_incident_count"] / zip_year_rates["n_months"]
)

# -----------------------------------------------------------------
# 5  Global scaling + KMeans
# -----------------------------------------------------------------
scaler = StandardScaler()
rates_scaled = scaler.fit_transform(zip_year_rates[["rate"]])

kmeans_global = KMeans(n_clusters=3, random_state=0, n_init=10)
zip_year_rates["cluster"] = kmeans_global.fit_predict(rates_scaled)

centroids_raw = (
    scaler.mean_[0] + kmeans_global.cluster_centers_.flatten() * scaler.scale_[0]
)
ordering = np.argsort(centroids_raw)  # low→high
label_map = {old: new for new, old in enumerate(ordering)}
zip_year_rates["congestion_level"] = zip_year_rates["cluster"].map(label_map)

# -----------------------------------------------------------------
# 6  Diagnostics
# -----------------------------------------------------------------
print("\n=== Global centroids (incidents / month) ===")
for i, c in enumerate(np.sort(centroids_raw)):
    print(f"Cluster {i}: {c:.3f}")
boundaries = (np.sort(centroids_raw)[:-1] + np.sort(centroids_raw)[1:]) / 2
print("\n=== Boundaries used for every ZIP ===")
print(f"Low / Medium : {boundaries[0]:.3f}")
print(f"Medium / High: {boundaries[1]:.3f}")
print("\n=== ZIP counts per cluster & year ===")
print(
    zip_year_rates.groupby(["Year", "congestion_level"])
    .size()
    .unstack(fill_value=0)
)

# -----------------------------------------------------------------
# 7  Build per‑year GeoDataFrames + scatter sources
# -----------------------------------------------------------------
geo_by_year, zip_counts_by_year = {}, {}
for yr in sorted(zip_year_rates["Year"].unique()):
    yr_slice = zip_year_rates[zip_year_rates["Year"] == yr].copy()
    zip_counts_by_year[yr] = yr_slice.rename(columns={"Incident Zip": "Incident_Zip"})
    geo_merged = geo_gdf.merge(
        yr_slice, left_on="MODZCTA", right_on="Incident Zip", how="left"
    )
    geo_merged["congestion_level"] = geo_merged["congestion_level"].fillna(0)
    geo_by_year[yr] = geo_merged

# -----------------------------------------------------------------
# 8  Plot
# -----------------------------------------------------------------
years = list(geo_by_year.keys())
num_cols = len(years)
if num_cols == 0:
    raise ValueError("No data after preprocessing.")

# Discrete palette to keep colours fixed (Viridis endpoints + mid)
palette = {0: "#440154", 1: "#21908C", 2: "#FDE725"}

# Common y‑axis range across all years
ymax = zip_year_rates["rate"].max() * 1.05

fig = make_subplots(
    rows=2,
    cols=num_cols,
    subplot_titles=[f"{y}<br>Map" for y in years]
    + [f"{y}<br>Scatter" for y in years],
    specs=[[{"type": "mapbox"}] * num_cols, [{"type": "xy"}] * num_cols],
    vertical_spacing=0.10,
    horizontal_spacing=0.05,
)

mapbox_style = "carto-positron"
center_nyc = dict(lat=40.7128, lon=-74.0060)
zoom_nyc = 9

for i, yr in enumerate(years, start=1):
    gdf = geo_by_year[yr]
    zdf = zip_counts_by_year[yr]

    # --- Map ---
    fig.add_trace(
        go.Choroplethmapbox(
            geojson=gdf.__geo_interface__,
            locations=gdf.index,
            z=gdf["congestion_level"],
            colorscale=[
                [0.00, palette[0]],
                [0.33, palette[0]],
                [0.34, palette[1]],
                [0.66, palette[1]],
                [0.67, palette[2]],
                [1.00, palette[2]],
            ],
            zmin=0,
            zmax=2,
            showscale=False,
            marker_opacity=0.7,
            marker_line_width=0,
            hovertemplate="<b>ZIP:</b> %{customdata}<br>"
            "<b>Level:</b> %{z}<extra></extra>",
            customdata=gdf["MODZCTA"],
            name=f"Map {yr}",
        ),
        row=1,
        col=i,
    )

    # --- Scatter ---
    fig.add_trace(
        go.Scattergl(
            x=zdf["Incident_Zip"],
            y=zdf["rate"],
            mode="markers",
            marker=dict(
                color=zdf["congestion_level"].map(palette),
                size=6,
            ),
            hovertemplate="<b>ZIP:</b> %{x}<br>"
            "<b>Rate:</b> %{y:.2f}<br>"
            "<b>Level:</b> %{marker.color}<extra></extra>",
            name=f"Scatter {yr}",
        ),
        row=2,
        col=i,
    )

# Layout
fig.update_layout(
    height=900,
    width=num_cols * 340,
    title_text= '',
    mapbox=dict(style=mapbox_style, center=center_nyc, zoom=zoom_nyc),
    **{
        f"mapbox{i}": dict(
            style=mapbox_style,
            center=center_nyc,
            zoom=zoom_nyc,
            domain={"x": [(i - 1) / num_cols, i / num_cols], "y": [0.50, 1.0]},
        )
        for i in range(1, num_cols + 1)
    },
    **{
        f"xaxis{i}": dict(title="ZIP Code", showgrid=True)
        for i in range(1, num_cols + 1)
    },
    **{
        f"yaxis{i}": dict(
            title="Incidents per Month (Jan–Apr)", showgrid=True, range=[0, ymax]
        )
        for i in range(1, num_cols + 1)
    },
    margin=dict(l=0, r=0, t=60, b=0),
)

# Discrete legend
for lvl, colour in palette.items():
    fig.add_trace(
        go.Scattergl(
            x=[None],
            y=[None],
            mode="markers",
            marker=dict(size=8, color=colour),
            legendgroup=f"level{lvl}",
            showlegend=True,
            name=["Low", "Medium", "High"][lvl],
        )
    )

fig.show()
```

```python
# ────────────────────────────────────────────────────────────────
# YoY change Jan‑Apr (2024→2025) per ZIP + K‑means clustering
# ────────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1 ── keep Jan‑Apr 2024 & 2025 only
df_yoy = (
    df_con
    .query("Year in [2024, 2025] and Month in [1, 2, 3, 4]")
    .copy()
)

# 2 ── incident counts by ZIP / Year / Month
counts = (
    df_yoy
    .groupby(["Incident Zip", "Year", "Month"])
    .size()
    .reset_index(name="incident_count")
)

# 3 ── wide format: one column per Year
counts_wide = (
    counts
    .pivot_table(index=["Incident Zip", "Month"],
                 columns="Year",
                 values="incident_count",
                 fill_value=0)
    .reset_index()
)

# 4 ── YoY change  (handle div‑by‑zero → NaN)
counts_wide["yoy_change"] = np.where(
    counts_wide[2024] == 0,
    np.nan,                               # undefined change
    counts_wide[2025] / counts_wide[2024] - 1
)

# 5 ── matrix: rows = ZIP, columns = months
zip_yoy = (
    counts_wide
    .pivot(index="Incident Zip", columns="Month", values="yoy_change")
    .rename(columns={1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr"})
)

# optional: replace inf / NaN (ZIPs with zero in 2024) with 0
zip_yoy = (
    zip_yoy
    .replace([np.inf, -np.inf], np.nan)
    .fillna(0)
)

# 6 ── K‑means clustering on the 4‑month YoY vectors
scaler = StandardScaler()
X      = scaler.fit_transform(zip_yoy)

kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
zip_yoy["cluster"] = kmeans.fit_predict(X)

# 7 ── see the result
print(zip_yoy.head())
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import plotly.graph_objects as go
from shapely.wkt import loads

# ────────────────────────────────────────────────
# 1.  INPUTS  (assume df_con already in memory)
# ────────────────────────────────────────────────
GEO_PATH = "Modified_Zip_Code_Tabulation_Areas__MODZCTA__20250427.csv"

# Geo shapes
geo = pd.read_csv(GEO_PATH)
geo["geometry"] = geo["the_geom"].apply(loads)
geo_gdf = gpd.GeoDataFrame(geo, geometry="geometry", crs="EPSG:4326")
geo_gdf["MODZCTA"] = pd.to_numeric(geo_gdf["MODZCTA"], errors="coerce")

# ────────────────────────────────────────────────
# 2.  FILTER 2024‑2025  (Jan–Apr)
# ────────────────────────────────────────────────
df = (
    df_con
    .query("Year in [2024, 2025] and Month in [1,2,3,4]")
    .copy()
)

# counts per (ZIP, Year, Month)
counts = (
    df.groupby(["Incident Zip", "Year", "Month"])
    .size()
    .reset_index(name="cnt")
)

# wide Year view
wide = counts.pivot_table(index=["Incident Zip", "Month"],
                          columns="Year", values="cnt", fill_value=0).reset_index()

# YoY change
wide["yoy"] = np.where(
    wide[2024] == 0,
    np.nan,
    wide[2025] / wide[2024] - 1
)

# restrict crazy values (optional cap to ±20 for readability)
wide["yoy_clipped"] = wide["yoy"].clip(-20, 20)

# average YoY over Jan‑Apr per ZIP
zip_yoy = (
    wide.pivot(index="Incident Zip", columns="Month", values="yoy_clipped")
    .mean(axis=1)
    .reset_index(name="mean_yoy")
)

# cluster by sign: decrease, no‑change, increase
def sign_cluster(v):
    if pd.isna(v):
        return 1   # treat undefined as no change
    if v < 0:  return 0   # decrease
    if v == 0: return 1   # no change
    return 2              # increase

zip_yoy["cluster"] = zip_yoy["mean_yoy"].apply(sign_cluster)

# merge into geo shapes
geo_merged = geo_gdf.merge(
    zip_yoy, left_on="MODZCTA", right_on="Incident Zip", how="left"
)
geo_merged["cluster"] = geo_merged["cluster"].fillna(1)

# ────────────────────────────────────────────────
# 3.  PLOTLY MAP
# ────────────────────────────────────────────────
palette = {0: "#440154", 1: "#21908C", 2: "#FDE725"}  # purplish, teal, yellow

# Build colour vector
geo_merged["color"] = geo_merged["cluster"].map(palette)

fig = go.Figure(go.Choroplethmapbox(
    geojson=geo_merged.__geo_interface__,
    locations=geo_merged.index,
    z=geo_merged["cluster"],
    colorscale=[[0, palette[0]], [0.33, palette[0]],
                [0.34, palette[1]], [0.66, palette[1]],
                [0.67, palette[2]], [1, palette[2]]],
    zmin=0, zmax=2,
    marker_opacity=0.7, marker_line_width=0,
    hovertemplate="<b>ZIP:</b> %{customdata}<br>"
                  "<b>YoY change:</b> %{text:.1%}<extra></extra>",
    customdata=geo_merged["MODZCTA"],
    text=geo_merged["mean_yoy"],
    showscale=False
))

fig.update_layout(
    mapbox=dict(
        style="carto-positron",
        center=dict(lat=40.7128, lon=-74.0060), zoom=9
    ),
    title="Traffic‑Complaint YoY Change (Jan‑Apr 2024 → 2025)<br>"
          "0 = Decrease, 1 = No Change, 2 = Increase",
    margin=dict(l=0, r=0, t=60, b=0)
)

# add manual legend
for lvl, label in zip([0,1,2], ["Decrease", "No change", "Increase"]):
    fig.add_trace(go.Scattermapbox(
        lat=[None], lon=[None],
        mode="markers",
        marker=dict(size=10, color=palette[lvl]),
        legendgroup=f"c{lvl}", showlegend=True, name=label
    ))

fig.show()
```

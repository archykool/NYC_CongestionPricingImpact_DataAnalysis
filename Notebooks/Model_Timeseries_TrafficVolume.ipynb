```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# 确保 Date 是 datetime 类型
aggregated_traffic["Date"] = pd.to_datetime(aggregated_traffic["Date"])

# 1. 计算每天的总流量
daily_total = aggregated_traffic.groupby("Date")["Vol"].sum().rename("total_vol")

# 2. 计算每天 peakhour 的流量
daily_peak = aggregated_traffic[aggregated_traffic["Time_Label"] == "peakhour"]
daily_peak = daily_peak.groupby("Date")["Vol"].sum().rename("peak_vol")

# 3. 合并并计算 peakhour 占比
daily_df = pd.concat([daily_total, daily_peak], axis=1).fillna(0)
daily_df["peak_ratio"] = daily_df["peak_vol"] / daily_df["total_vol"]

# 4. 设置 Date 为索引并按天频率填补缺失
daily_df = daily_df.asfreq("D").fillna(0)

# 5. 对总交通量进行季节性分解
decomp_total = seasonal_decompose(daily_df["total_vol"], model="additive", period=7)
decomp_ratio = seasonal_decompose(daily_df["peak_ratio"], model="additive", period=7)

# 6. 可视化 - 总量季节性
plt.figure(figsize=(12, 8))
decomp_total.plot()
plt.suptitle("Daily Total Traffic Volume - Seasonal Decomposition", fontsize=14)
plt.tight_layout()
plt.show()

# 7. 可视化 - 占比季节性
plt.figure(figsize=(12, 8))
decomp_ratio.plot()
plt.suptitle("Daily Peakhour Ratio - Seasonal Decomposition", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error

# 日期字段转为 datetime 类型
aggregated_df['Date'] = pd.to_datetime(aggregated_df['Date'])

# 筛选高峰时段数据
daytime_df = aggregated_df[aggregated_df['Time_Label'] == 'peakhour']

# 白天通行比例
vehicle_class_daytime = daytime_df.groupby(['Date', 'Vehicle Class'])['CRZ Entries'].sum().reset_index()
vehicle_class_all_periods = aggregated_df.groupby(['Date', 'Vehicle Class'])['CRZ Entries'].sum().reset_index()

merged_df = pd.merge(vehicle_class_daytime, vehicle_class_all_periods,
                     on=['Date', 'Vehicle Class'], suffixes=('_daytime', '_all'))
merged_df['Daytime Ratio'] = merged_df['CRZ Entries_daytime'] / merged_df['CRZ Entries_all']
merged_df = merged_df.replace([np.inf, -np.inf], np.nan).dropna(subset=['Daytime Ratio'])

# 建模与预测
vehicle_classes = merged_df['Vehicle Class'].unique()

for vehicle in vehicle_classes:
    print(f"\nAnalyzing Vehicle Class: {vehicle}")
    
    vehicle_data = merged_df[merged_df['Vehicle Class'] == vehicle]

    if len(vehicle_data) < 14:
        print(f"Not enough data points for {vehicle} (less than 14). Skipping.")
        continue

    vehicle_data.set_index('Date', inplace=True)
    vehicle_data = vehicle_data[['Daytime Ratio']].sort_index()

    train_size = int(len(vehicle_data) * 0.8)
    train, test = vehicle_data[:train_size], vehicle_data[train_size:]

    try:
        print(f"Running Seasonal SARIMAX for {vehicle}")

        # 构建 SARIMAX 模型（相当于 ARIMA(1,1,1)(1,1,1,7)）
        model = SARIMAX(train, 
                        order=(1, 1, 1), 
                        seasonal_order=(1, 1, 1, 7),
                        enforce_stationarity=False,
                        enforce_invertibility=False)
        result = model.fit(disp=False)

        # 预测
        forecast = result.predict(start=test.index[0], end=test.index[-1], dynamic=False)
        
        mape = mean_absolute_percentage_error(test, forecast)
        rmse = np.sqrt(mean_squared_error(test, forecast))

        print(f"MAPE for {vehicle}: {mape:.4f}")
        print(f"RMSE for {vehicle}: {rmse:.4f}")

        # 可视化
        plt.figure(figsize=(10, 6))
        plt.plot(train.index, train['Daytime Ratio'], label='Train')
        plt.plot(test.index, test['Daytime Ratio'], label='Test', color='blue')
        plt.plot(forecast.index, forecast, label='Predicted', color='red')
        plt.title(f"SARIMAX Forecast for Daytime Driving Ratio ({vehicle})")
        plt.xlabel('Date')
        plt.ylabel('Daytime Ratio')
        plt.legend()
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error fitting SARIMAX model for {vehicle}: {e}")
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress

# Store summary results
trend_summary = []

# Loop through each vehicle class
for vehicle in merged_df['Vehicle Class'].unique():
    vehicle_data = merged_df[merged_df['Vehicle Class'] == vehicle].copy()
    vehicle_data = vehicle_data.sort_values('Date')
    
    if len(vehicle_data) < 10:
        trend_summary.append((vehicle, 'Insufficient data', None))
        continue

    # Convert date to numerical value (days since the first date)
    vehicle_data['Days'] = (vehicle_data['Date'] - vehicle_data['Date'].min()).dt.days
    X = vehicle_data[['Days']]
    y = vehicle_data['Daytime Ratio']
    
    # Fit linear regression model
    reg = LinearRegression()
    reg.fit(X, y)
    slope = reg.coef_[0]

    # Get statistical info using linregress
    stats = linregress(vehicle_data['Days'], vehicle_data['Daytime Ratio'])
    
    # Determine trend direction
    if stats.pvalue < 0.05:
        if slope > 0.0005:
            trend = 'Significant upward trend'
        elif slope < -0.0005:
            trend = 'Significant downward trend'
        else:
            trend = 'Slight variation (approximately stable)'
    else:
        trend = 'No significant trend'

    # Save results
    trend_summary.append((vehicle, trend, slope))
    
    # Plot trend
    plt.figure(figsize=(8, 5))
    plt.plot(vehicle_data['Date'], vehicle_data['Daytime Ratio'], label='Actual')
    plt.plot(vehicle_data['Date'], reg.predict(X), label='Trend Line', color='red')
    plt.title(f'Trend in Daytime Driving Ratio - {vehicle}')
    plt.xlabel('Date')
    plt.ylabel('Daytime Ratio')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Print the summary report
print("\n📋 Trend Analysis Report:")
for vehicle, trend, slope in trend_summary:
    if slope is not None:
        print(f"- {vehicle}: {trend} (approx. daily change: {slope:.4f})")
    else:
        print(f"- {vehicle}: {trend}")
```

```python
import folium
from folium.plugins import TimestampedGeoJson
import pandas as pd

# 降采样（每个桥每周）
grouped_df['Date'] = pd.to_datetime(grouped_df['Date'])
grouped_df = grouped_df.sort_values(['Detection Group', 'Date'])
sampled_df = grouped_df.groupby('Detection Group').resample('7D', on='Date').mean(numeric_only=True).reset_index()

# 为每个桥计算自己的分位数（33%、66%）
bridge_thresholds = sampled_df.groupby("Detection Group")["CRZ Entries"].quantile([0.33, 0.66]).unstack()

# 设置颜色函数（基于各桥自己的分布）
def get_bridge_color(bridge, vol):
    low, high = bridge_thresholds.loc[bridge]
    if vol > high:
        return "red"
    elif vol > low:
        return "orange"
    else:
        return "green"

# 构建 GeoJSON 数据
features = []
for _, row in sampled_df.iterrows():
    bridge = row["Detection Group"]
    volume = row["CRZ Entries"]
    color = get_bridge_color(bridge, volume)
    radius = max(volume / 1000, 3)
    
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row["Longitude"], row["Latitude"]],
        },
        "properties": {
            "time": row["Date"].strftime("%Y-%m-%d"),
            "popup": f"<b>{bridge}</b><br>Volume: {int(volume):,}",
            "icon": "circle",
            "iconstyle": {
                "fillColor": color,
                "fillOpacity": 0.7,
                "stroke": "true",
                "strokeColor": "#000",
                "strokeOpacity": 0.4,
                "radius": radius
            }
        }
    }
    features.append(feature)

geojson = {
    "type": "FeatureCollection",
    "features": features
}

# 创建地图
m = folium.Map(location=[40.71, -74], zoom_start=12, tiles="CartoDB positron")

# 添加时间滑块图层
TimestampedGeoJson(
    geojson,
    period="P7D",  # 每周一帧，速度更快
    add_last_point=False,
    auto_play=True,
    loop=False,
    max_speed=3,
    loop_button=True,
    date_options="YYYY-MM-DD",
    time_slider_drag_update=True
).add_to(m)

# 图例（说明为：颜色依据每座桥自身的流量区间）
legend_html = """
<div style="
    position: fixed; 
    bottom: 50px; left: 50px; width: 220px; height: 130px; 
    background-color: white; z-index:9999; 
    font-size:14px; border:2px solid gray; padding:10px;">
    <b>Traffic Volume Legend</b><br>
    <i style="background:red; width:10px; height:10px; display:inline-block;"></i> High (Top 33%)<br>
    <i style="background:orange; width:10px; height:10px; display:inline-block;"></i> Medium<br>
    <i style="background:green; width:10px; height:10px; display:inline-block;"></i> Low (Bottom 33%)<br>
    <small>* Thresholds by bridge</small>
</div>
"""
m.get_root().html.add_child(folium.Element(legend_html))

m.save("bridge_traffic_time_heatmap_customized.html")
print("✅ 优化速度并自适应颜色的地图已保存为 bridge_traffic_time_heatmap_customized.html")
```
